# Data Cleaning

After importing the data, we need to clean it so that it can be used for analysis. Since we are specifically looking for the trends of common birds in the forest plots, we can drop the columns we don't need. We will also fix any incorrect datatypes.

```python
# Convert Date to a date type, as it is currently an object
from datetime import date

df['Date']= pd.to_datetime(df['Date'])
df.info()
```

```python
# Filter data to only be of forests

df = df[df['Location_Type'] == 'Forest']
df.info()
```

```python
# Filter out unneeded columns
col_to_keep = [
    'Plot_Name',
    'Year',
    'Date',
    'ID_Method',
    'Common_Name',
    'Scientific_Name'
]

df = df[col_to_keep]
df.info()
```

## Missing Values

The data is nearly complete for analysis, but there are still 319 missing values in `ID_Method`, and 1 missing value in `Common_Name` and `Scientific_Name`. We can drop the single missing value in the two name columns, and replace missing values in `ID_Method` with "Unknown" so they can still be included in the analysis without affecting the trends.

```python
# Drop single row with missing name
df = df.dropna(subset=['Common_Name', 'Scientific_Name'])

# Replace missing ID_Method values with 'Unknown'
df['ID_Method'] = df['ID_Method'].fillna('Unknown')

df.info()
```

# Results

Overall, the top 5 most observed birds in the forests of Manassas National Battlefield Park are:
- Blue-gray Gnatcatcher
- Carolina Wren
- Eastern Tufted Titmouse
- Northern Cardinal
- Red-eyed Vireo

In terms of the yearly trends for these birds, we can see that observations for the Northern Cardinal dropped the most over a 10-year span, steadily going down despite having a decent increase from 2011 to 2012.

The Carolina Wren experienced the largest drop in observations from 2009 to 2010, but began to grow, spiking in 2013 before dropping once again.

The Eastern Tufted Titmouse also experienced a notable spike of observations in 2013, which also declined in subsequent years.
